```python
import pandas as pd
import json
df=pd.DataFrame(columns=['时间','搜索量','产品数','均价','搜索转化率','品牌数量'])
```

```python
fp = open("amz_trends.txt", 'r')
f=fp.read()
s=f.replace(' ','').replace('\n','')
js=json.loads(s)
ss=js['data']['niche']['trendsMetrics']

# for i in ss:
#     

# df
ss



```

```python
bb=[]
for i in ss:
    aa=[i['datasetDate'],i['searchVolumeT7'],i['productCount'],i['averagePriceT7'],i['searchConversionRateT7'],i['brandCount']]
    bb.append(aa)

df=pd.DataFrame(bb,columns=['时间','搜索量','产品数','均价','搜索转化率','品牌数量'])
df['预估销量']=df['搜索量']*df['搜索转化率']
df
```

```python
import matplotlib.pyplot as plt
df['预估销量'].plot(kind='line',figsize=(50,20),alpha=1)
for m,n in zip (df.index,df['预估销量']):
    plt.text(m,n,n,ha='center',va='bottom',fontsize=10)
plt.savefig('pci.png')
plt.show()
```
