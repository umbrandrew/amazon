```python
import pandas as pd
import csv
path_list=[3,4,5]
tt=pd.DataFrame()
for n in path_list:
    f = open('亚马逊%d月.csv'%n,'r',encoding='utf-8')
    reader = csv.reader(f)
    t1 = pd.DataFrame(reader,dtype=str)
    t1=t1.iloc[7:]
    t1.columns=t1.iloc[0]
    t1=t1.reset_index(drop=True)
    tt=pd.concat([t1,tt])
    
tt=tt.drop(tt[tt['date/time']=='date/time'].index)



```

```python
tt_order_info=tt[tt['type']=='Order']
tt_order_info=tt_order_info[['date/time', 'order id','sku','quantity','fulfillment','order city','order state','order postal','product sales','product sales tax','shipping credits','shipping credits tax','gift wrap credits','giftwrap credits tax','Regulatory Fee','Tax On Regulatory Fee','promotional rebates','promotional rebates tax','marketplace withheld tax','selling fees','fba fees','other transaction fees','other','total']]
ss=tt_order_info['date/time'].str.split(' ', expand=True)
tt_order_info.drop('date/time',axis=1,inplace=True)
tt_order_info.insert(0,'日期',ss[2].str.cat([ss[0],ss[1].str[:-1]],sep='-',na_rep='?'))
tt_order_info.insert(1,'时间',ss[3])
tt_order_info.info()
```
